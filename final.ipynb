```python
time = ds.variables['time'][:]
lat = ds.variables['latitude'][:]
lon = ds.variables['longitude'][:]
sla = ds.variables['sla']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.pcolor(lon,lat, sla[0,:,:],cmap='winter')
plt.title('DT merged all satellites Global Ocean Gridded SSALTO/DUACS Sea Surface Height L4 product and derived variables')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar()
plt.show()
```

```python
def realday(a):
    import datetime
    b = int(time[a])
    date = datetime.date(1950,1,1)
    c = datetime.timedelta(days = b)
    realdate = date + c 
    return realdate
```
